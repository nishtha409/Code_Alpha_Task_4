```python
import pandas as pd
data=pd.read_csv('titanic.csv')
print(data.head())
```

```python
#Checking null values in the original Dataset
print("Is null")
print(data.isnull())
```

```python
# cleaning the data can be done through drop and fill
data_temp=data
#describe() for original datasets
data.describe()
data_temp.dropna(axis=0)
#data info() for Original Datasets
# data.info()
#DROPPING NULL DATA
#axis=0 then selecting rows & axis=1 then selecting column
```

```python
#FILLING null values of Age Column
data_temp["Age"]=data_temp["Age"].fillna(30)
print(data_temp)
```

```python
#outliers
import seaborn as sns
sns.boxplot(data["Age"])
```

```python
removed_outliers=data[data["Age"]<65]
sns.boxplot(removed_outliers["Age"])
```

```python
# duplicate
data = data.sort_values("Name")
dup = data["Name"].duplicated()
data[dup]
# compand to remove duplicate
data = data.drop_duplicates()
```

```python
data.head()
```

```python
#Dropping NaN values from the table before Normalization
data=data.dropna()
```

```python
# data transformation
# implementing the scaling for normalizing the data
from sklearn.preprocessing import normalize
data = data.dropna()
data_age = data["Age"]
data_age = data_age.values.reshape(-1,1)
data_fare = data["Fare"]
data_fare = data_fare.values.reshape(-1,1)
data_age_norm = normalize(data_age)
data_fare_norm = normalize(data_fare)
data["Age"] = data_age_norm
data["Fare"] = data_fare_norm
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
data = data.dropna()
data_age = data['Age'].values.reshape(-1,1)
scale = MinMaxScaler()
scale = scale.fit(data_age)
data_age_norm = scale.transform(data_age)
data['Age'] = data_age_norm
data.head()
```

```python
#Convert a categorical variable into numerical representation (LabelEncode --is used when there is--Binary
#features in the column / 2 or 3 different feature ):
#columns remain same, in the same column the values are replaced by 1 or 0
#male=0
#female=1
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_sex = data['Sex']
data['Sex'] = le.fit_transform(data_sex)
data['Sex'].unique()
data.head()
```
